#Data Encoding

if we have :<br>
2 categories: Presence -> 1, Absence -> 0<br>
5 categories: 1, 2, 3, 4, 5 -> 1 to n or 0 to n-1 -> Label Encoding: We can only use it for target variables(Variables that need to be predicted). We cannot use it on predictor variables <br>

Because when we write them as numbers like 1,2,3,4,5 ,there is an order in them. 5 is greater than 1. When we multiply a value by 5 or add a value 5 it is obviously going to have a greater effect on the result rather than the value 1<br>

Suppose we take 2<br>
- we multiply 1 to it
- we multiply 5 to it

When do we get a greater result?<br>
The second case (2 x 5=10)<br>

If there is an ordering in this way, some categories get more preference than the others, which we do not want to happen<br>

So what we do is<br>
For 5 categories we create 5 more columns and for each column we consider values as 0 and 1<br>
Category 1: Presence -> 1, Absence -> 0<br>
Category 2: Presence -> 1, Absence -> 0<br>
Category 3: Presence -> 1, Absence -> 0<br>
Category 4: Presence -> 1, Absence -> 0<br>
Category 5: Presence -> 1, Absence -> 0<br>

This process is called -> One Hot Encoding<br>

#Data Scaling

Problem: Predict the price of house based on its features

We have two features<br>
Area of the house(in sq ft), No of bedrooms<br>
- Area: 10000, 20000, 15000, etc<br>
- No of bedrooms: 1,2,3,etc.<br>

Algorithms are mainly mathematical, if we multiply a number by 10000 and if we multiply a number by 1, obviously 10000 will have a larger influence<br>

- 2 x 10000=20000
- 2 x 1=2

So the model will give more preference to the area variable just because it has higher value<br>

The model might think that the area variable is more important than the number of bedrooms <br>

So we change the value of the variables to the same range<br>


```
data[num].describe()
```

#Standard Scalar

(X-Mean(X))/Std Dev(X)

formula is similar to that of Z distribution

- widely used

```
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```
std_scaled_data = scalar.fit_transform(data[num])
```

```
pd.DataFrame(std_scaled_data, columns=num).describe()
```

```
#After running standard scalar
#Mean is approx 0
#Std is approx 1
```

#Min Max Scalar

[Max(X)-X]/[Max(X)-Min(X)]

X is Max => 0<br>
X is Min => 1<br>


- Used if there are variables which shouldn't have negative values
- If we want a probability(always lying between 0 and 1) or a proper fraction

```
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```
min_max_scaled_data = scaler.fit_transform(data[num])
```

```
pd.DataFrame(min_max_scaled_data, columns=num).describe()
```
